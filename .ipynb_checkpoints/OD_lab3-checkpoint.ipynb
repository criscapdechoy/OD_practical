```python
result.fillna('None',inplace=True)
sum(result.venue=='None')
```

```python
result.fillna('None',inplace=True)
result['publishedIn'] = result['pub_type']
result.rename(columns={'pub_type':'subclassOf'}, inplace=True)
result.drop(result[result['doi'].str.contains('::')].index,axis=0,inplace=True)
result.drop(result[result['subclassOf'].str.contains('ook')].index,axis=0,inplace=True)
```

```python
result.loc[result['venue'].str.contains("CIKM"),'publishedIn'] = 'Proceedings'
result.loc[result['venue'].str.contains("SIGIR"),'publishedIn'] = 'Proceedings'
result.loc[result['venue'].str.contains("WWW"),'publishedIn'] = 'Proceedings'
result.loc[result['venue'].str.contains("CSCW"),'publishedIn'] = 'Proceedings'
result.loc[result['venue'].str.startswith("J."),'publishedIn'] = 'Journal'
```

```python
result.loc[result['venue'].str.contains("onference"),'publishedIn']='Proceedings'
result.loc[result['subclassOf'].str.contains("onference"),'publishedIn']='Proceedings'
result.loc[result['publishedIn'].str.contains("oceedings"),'subclassOf']='Proceeding Paper'
result.loc[result['subclassOf'].str.contains("ournal"),'publishedIn']='Journal'
result.loc[result['publishedIn'].str.contains("ournal"),'subclassOf']='Article'
result.loc[result['publishedIn'].str.contains("None"),'publishedIn']='Proceedings'
```

```python
from numpy import nan
j, i = 0,0 
edition = list()
city = list()
volume = list()
issue = list()
reviewed = list()
paper_prop = list()
paper_id = list()
pages = list()


def only_upper(s):
    return "".join(c for c in s if c.isupper())


type_paper = ['DemoPaper', 'FullPaper', 'ShortPaper', 'SurveyPaper']
cities = ['Beijing', 'Berlin','Brussels','Buenos Aires','Copenhagen',
          'Istanbul','Paris','Barcelona','London','Seoul','Singapore',
          'Sydney','Tokyo']

for i,cols in result[['venue','publishedIn','year','vol','pages']].iterrows():
    paper_id.append(f"Paper_{i}")
    if cols.publishedIn == 'Proceedings':
        venue = cols.venue.replace('th','').replace('rd','').replace('st','').replace('nd','')
        numbers = [int(s) for s in re.findall(r'\b\d+\b', str(venue))]
        if numbers:
            edition.append(sorted(numbers)[0]) 
            i += 1
        else:
            j += 1
            edition.append(cols.year)
        city.append(cities[random.randint(0,len(cities),1)[0]])
    else:
        edition.append(nan)
        city.append(nan)
        
    if cols.publishedIn == 'Journal':
        
        try:
            volume.append(int(vol))
        except:
            volume.append(int(int(abs(random.randn(1)[0]*10))))
        page0 = int(abs(random.randn(1)[0]*10)) 
        page1 = f"{page0}:{int(page0 + 12)}" 
        pages.append(page1)
        issue.append(int(abs(random.randn(1)[0]*25)))
    else:
        volume.append(nan)
        pages.append(nan)
        issue.append(nan)
    ind = int(abs(random.randn(1)[0]*len(unique_authors)*0.25))
    
    try:
        reviewed.append(unique_authors.iloc[ind][0])
    except:
        reviewed.append('None')
    paper_prop.append(type_paper[random.randint(0,3,1)[0]])

result['paper_id'] = paper_id
result['paper_prop']= paper_prop
result['edition'] = edition
result['heldIn'] = city
result['reviewedBy'] = reviewed
result['vol'] = volume
result['issue'] = issue
result['pages'] = pages
result['venue'] = result['venue'].apply(lambda c: only_upper(c))
result['pub_type_type'] = result['publishedIn'].apply(lambda c: "OpenAccessJournal" if c.find("ournal")!=-1 else "ConferenceProceedings")

result.head()#[int(s) for s in re.findall(r'\b\d+\b', x)]
```

```python
affil = list()
publications = list()
index_affiliation = random.randint(0,len(affiliations),len(unique_authors))
for i, cols in authors.iterrows():
    index = index_affiliation[unique_authors[0]==cols.author_name][0]
    affil.append(affiliations.iloc[index_affiliation[index]][0])
authors['affiliations'] = affil
```

```python
final_result = pd.merge(authors,result,on='doi')
```

```python
import csv
from rdflib import Graph, Literal, Namespace, URIRef
from rdflib.namespace import DCTERMS, RDF, RDFS, SKOS, XSD


input_file = csv.DictReader(open("test_sheet.csv"))


# make a graph
output_graph = Graph()


for row in input_file:

	# convert it from an OrderedDict to a regular dict
	row = dict(row)

	#{'Subject Label': 'Pearl Wilmer Booker', 'Subject URI': 'None', 'Predicate Label': 'Daughter Of', 'Predicate URI': '', 'Predicate Symmetry': 'Asymmetric', 'Object Label': 'Mary Booker', 'Object URI': 'None'}
	# make a literal and add it
	output_graph.add(  (URIRef(row['Subject URI']), RDFS.label, Literal(row['Subject Label'], lang='en')) )

	# make a triple with the object as uri
	output_graph.add(  (URIRef(row['Subject URI']), URIRef(row['Predicate URI']), URIRef(row['Object URI'])) )	



output_graph.serialize(destination='my_graph.nt', format='nt')
```
